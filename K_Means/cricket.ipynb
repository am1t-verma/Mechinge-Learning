```python
from sklearn.preprocessing import StandardScaler
ss  = StandardScaler()
```

```python
x = ss.fit_transform(df)
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,20):
    k = KMeans(n_clusters=i, init='random')
    k.fit(x)
    wcss.append(k.inertia_)

wcss
```

```python
plt.plot(wcss)
```

```python
# n_cluster = 15
```

```python
model = KMeans(n_clusters=5, init='random')
model.fit(x)
```

```python
df['lables'] = model.labels_
```

```python
df
```

```python
import plotly.express as px

fig = px.scatter_3d(df,
                    x='Mat',
                    y='Inns', 
                    z='Runs',
                    color='lables')
fig.show()
```
