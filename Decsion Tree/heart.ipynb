```python
tree_depth  = [1,2,3,4,5,6,7,8]
```

```python
acc_list = []
acc_test_list = []
acc_train_list = []
for i in tree_depth:
    model = DecisionTreeClassifier(random_state=23, max_depth = i)
    model.fit(x_train, y_train)
    y_pred_test = model.predict(x_test)
    y_pred_train = model.predict(x_train)

    acc_test = accuracy_score(y_pred_test, y_test)
    acc_test_list.append(acc_test)
    acc_train = accuracy_score(y_pred_train, y_train)
    acc_train_list.append(acc_train)

```

```python
pd.DataFrame({
    'Tree_Depth' : tree_depth,
    'Train_acc' : acc_train_list,
    'Test_acc' : acc_test_list,
})
```

# Hyperparameter Tuning

* This Is Control The Learning Process. 

* In Another Words Customization Before Traning.

* It Inc. The Accuracy Of Model.

* It Takes Less Time.

* Prevent OverFiting And UnderFiting.

### Grid Search CV

Why Use GridSearchCV?
To automatically test multiple hyperparameter combinations.

To select the best model based on a scoring metric (like accuracy, R², etc.).

It combines hyperparameter tuning + cross-validation in one step.

### What is K-Fold Cross-Validation?

How It Works:
The dataset is split into K equal parts (folds).

For each iteration (K total):

Use K-1 folds for training.

Use the remaining 1 fold for testing.

Repeat this process K times, so that each fold is used as the test set once.

```python
df.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_model = DecisionTreeClassifier()
```

```python
perameter = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [8, 9, 11, 12, 13],
    'random_state' : [42, 45, 64, 2, 41, 5],
    'min_samples_split' : [5,6,7,8,11],
    'min_samples_leaf' : [2,7,9]
}
```

```python
gv = GridSearchCV(estimator = hyper_model, param_grid = perameter, cv = 5)
```

```python
gv.fit(x_train, y_train)
```

```python
gv.get_params()
```

```python
gv.best_params_
```

```python
gv_model = gv.best_estimator_
```

```python
y_pred_gv = gv_model.predict(x_test)
```

```python
df['target'].value_counts()
```

```python
acc_gv_test = accuracy_score(y_pred_gv, y_test)
acc_gv_test
```

```python
y_pred_train_gv = gv_model.predict(x_train)
```

```python
acc_gv_train = accuracy_score(y_pred_train_gv, y_train)
```
